```python
from sklearn.manifold import TSNE
import pickle
embedding = []

for img in range(0, 24):
    with open("./" + str(img), "rb") as f:
        car_embedding = pickle.load(f)
        embedding.append(car_embedding)
    break
```

```python
import numpy as np
embedding = np.array(embedding)
```

```python
embedding = embedding.reshape([41, -1])
```

```python
tsne_embedded = TSNE(n_components=2).fit_transform(embedding)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20, 20
```

```python
tag = [0]
N = 1536
for i in range(5):
    tag += [i] * 8
    
tag = np.array(tag)

# define the colormap
cmap = plt.get_cmap('jet', 63)
# extract all colors from the .jet map
cmaplist = [cmap(i) for i in range(cmap.N)]
# create the new map
cmap = cmap.from_list('Custom cmap', cmaplist, cmap.N)

plt.scatter(tsne_embedded[:, 0], tsne_embedded[:, 1] ,c=tag ,cmap=cmap)
```
